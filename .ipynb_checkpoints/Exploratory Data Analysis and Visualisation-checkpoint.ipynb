```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid", color_codes=True)

import warnings
warnings.filterwarnings("ignore")

from pandas.plotting import radviz
```

```python
data = pd.read_csv('voice.csv')
```

### Checking first few data entries

```python
data.head()
```

### Check Dimentions of Data and Column Names

```python
print(data.shape) # 3168 data points with 21 variables/features
print()
print(data.columns)
```

```python
print(data.info()) # gives general information about data and check for null values
```

### Distribution of male female data

```python
sns.FacetGrid(data, hue="label", size=5) \
   .map(sns.kdeplot, "meanfun") \
   .add_legend()
plt.show()
```

```python
#Scatter Plot

sns.FacetGrid(data, hue="label", size=5).map(plt.scatter, "meanfun", "meanfreq").add_legend()
plt.show()
```

```python
# Boxplot
sns.boxplot(x="label",y="meanfun",data=data,hue='label')
plt.show()
```

```python
sns.swarmplot(x="label", y="meanfun", data=data, color=".15")
plt.show()
```

```python
# compare each feature
fig, ax = plt.subplots(figsize=(12,12))
radviz(data, "label")
plt.show()
```
